```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
faturamento = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

plt.figure(figsize=(10, 6))  # aumenta o tamanho da imagem

barras = plt.bar(lojas, faturamento, color='#008080', edgecolor='black')

# Título e eixos
plt.title("Faturamento por Loja", fontsize=16)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Faturamento (R$)", fontsize=12)

# Adiciona valores em cima das barras
for barra, valor in zip(barras, faturamento):
    plt.text(barra.get_x() + barra.get_width() / 2, valor + 20000,
             f"R$ {valor:,.0f}".replace(",", "X").replace(".", ",").replace("X", "."),
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Grid e layout
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Estilo
sns.set_style("whitegrid")

# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
produtos_mais = ['Micro-ondas', 'Livro: Programação', 'Kit Banquetas', 'Cama Box']
vendas_mais = [132, 145, 128, 151]

produtos_menos = ['Headset', 'Tabuleiro', 'Blocos Montar', 'Guitarra']
vendas_menos = [12, 18, 14, 16]

# Posicionamento das lojas no eixo Y
y = np.arange(len(lojas))
altura = 0.35

# Criar a figura
fig, ax = plt.subplots(figsize=(12, 6))

# Barras horizontais
ax.barh(y - altura/2, vendas_mais, height=altura, label='Mais Vendido', color='#4C72B0', edgecolor='black', linewidth=1.5)
ax.barh(y + altura/2, vendas_menos, height=altura, label='Menos Vendido', color='#DD8452', edgecolor='black', linewidth=1.5)

# Rótulos e valores nas barras
for i in range(len(lojas)):
    ax.text(vendas_mais[i] + 2, y[i] - altura/2, f'{produtos_mais[i]} ({vendas_mais[i]})', va='center', fontsize=10, color='#333')
    ax.text(vendas_menos[i] + 2, y[i] + altura/2, f'{produtos_menos[i]} ({vendas_menos[i]})', va='center', fontsize=10, color='#333')

# Estética e ajustes
ax.set_yticks(y)
ax.set_yticklabels(lojas, fontsize=12)
ax.set_xlabel('Unidades Vendidas', fontsize=13)
ax.set_title('Comparativo de Produtos Mais e Menos Vendidos por Loja', fontsize=16, weight='bold')

# Ajustando o limite do eixo X para melhorar a visualização
ax.set_xlim(0, max(vendas_mais) + 20)

# Adicionando grid para facilitar a leitura
ax.grid(True, linestyle='--', alpha=0.7)

# Movendo a legenda um pouco para baixo
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.02), fontsize=12)

# Ajuste final do layout para evitar sobreposição
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

```python
# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes = [34.69, 33.62, 33.07, 31.28]

# Ordenar os dados para exibir do maior para o menor
lojas_ordenadas = [x for _, x in sorted(zip(fretes, lojas), reverse=True)]
fretes_ordenados = sorted(fretes, reverse=True)

# Paleta de cores usando seaborn
cores = sns.color_palette("Blues_d", len(fretes))

# Criar gráfico
plt.figure(figsize=(10, 6))
bars = plt.barh(lojas_ordenadas, fretes_ordenados, color=cores)

# Adicionar valores ao final de cada barra
for bar in bars:
    valor = bar.get_width()
    plt.text(valor + 0.3, bar.get_y() + bar.get_height()/2,
             f'R$ {valor:.2f}', va='center', fontsize=11, weight='bold', color='#333')

# Destaques visuais
plt.title('Frete Médio por Loja', fontsize=16, weight='bold')
plt.xlabel('Frete Médio (R$)', fontsize=12)
plt.gca().invert_yaxis()  # Deixar a loja com maior valor no topo
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Estética
sns.despine(left=True, bottom=True)
plt.xlim(0, max(fretes) + 5)
plt.tight_layout()
plt.show()
```

# **Relatório Final: Análise de Desempenho das Lojas**

## Introdução

O objetivo deste relatório é ajudar o Senhor João a decidir qual loja ele deve vender, com base em uma análise dos dados de desempenho de cada uma das lojas. Alguns fatores importantes foram levados em consideração, como:

- O **faturamento total** de cada loja.
- As **categorias de produtos** mais e menos vendidas.
- A **média das avaliações** dos clientes por loja.
- Os **produtos mais e menos vendidos**.
- O **frete médio** de cada loja.

Através dessa análise, buscamos identificar qual loja tem o melhor desempenho, levando as consederações acima, além de outros pontos críticos que podem influenciar na escolha da loja a ser vendida.

---

## Desenvolvimento da Análise

### **1. Faturamento Total**

O faturamento total de cada loja é um fator fundamental para a análise, pois mostra o desempenho financeiro das lojas. Após analisar os dados, podemos observar que a **Loja 1** teve o maior faturamento, seguida pelas **Lojas 2, 3 e 4**.

 - **Loja 1**: R$ 1.534.509,12

 - **Loja 2**: R$ 1.488.459,06

 - **Loja 3**: R$ 1.464.025,03

 - **Loja 4**: R$ 1.384.497,58

O faturamento mais alto da **Loja 1** indica que ela teve um desempenho superior, o que pode ser um indicativo de que essa loja tem mais clientes ou vende produtos de maior valor. Esse fator pode ser um ponto a favor na hora de decidir qual loja vender, já que uma loja com maior faturamento tende a ser mais atraente para compradores.

### **2. Categorias de Produtos Mais e Menos Vendidas**

A análise das **categorias de produtos** mais e menos vendidas revelou que:

- **Loja 1** se destaca principalmente em **Eletrodomésticos** e **Eletrônicos**, produtos que possuem uma boa margem de lucro e uma alta demanda no mercado.
- **Loja 2** é forte em **Livros** e **Utilidades Domésticas**, que também são categorias populares, mas com menos variação de preços.
- **Loja 3** tem um bom desempenho em **Brinquedos** e **Esporte e Lazer**, categorias voltadas para nichos específicos, mas com potencial de crescimento, especialmente em épocas sazonais como a do Natal.
- **Loja 4** se destaca em **Móveis** e **Instrumentos Musicais**, o que sugere um foco mais em produtos pesados e de nicho, com menor giro, mas também com bom potencial de lucro.

Essas categorias mostram a diversidade do foco de venda de cada loja.

### **3. Média das Avaliações de Clientes**

Outro ponto muito importante é a **média das avaliações dos clientes** por loja. Isso pode indicar a qualidade do atendimento e a satisfação dos consumidores. A média das avaliações foi a seguinte:

 - **Loja 1**: 3.98

 - **Loja 2**: 4.04

 - **Loja 3**: 4.05

 - **Loja 4**: 3.99

A **Loja 3** teve a melhor média de avaliações, com 4.05, o que indica que os clientes estão satisfeitos com os produtos e serviços dessa loja. Em contrapartida, a **Loja 1**, embora tenha o maior faturamento, tem uma avaliação um pouco inferior, o que pode significar problemas de atendimento, de processos como entrega e transporte ou de qualidade de produto.

As boas avaliações da **Loja 3** podem ser um bom indicativo de uma base de clientes fiel e satisfeita, o que é um ponto positivo caso o Senhor João queira vender uma loja que já tenha boa reputação no mercado.

### **4. Produtos Mais e Menos Vendidos**

Em relação aos **produtos mais e menos vendidos**, foi observado que a **Loja 1** tem produtos como **micro-ondas** e **cama box** entre os mais vendidos, enquanto produtos como **guitarra** e **headset** não tiveram tantas vendas. Já a **Loja 3** teve maior sucesso com itens como **brinquedos** e **kits de esporte**, o que reflete uma base de clientes diferente, talvez mais focada em produtos de lazer e diversão.

A diversidade nos produtos vendidos também pode influenciar na decisão. Produtos com grande volume de vendas e boa aceitação podem garantir uma base sólida de receita para a loja, enquanto produtos de nicho podem representar uma oportunidade ou risco, dependendo do perfil de mercado.

### **5. Frete Médio**

Por fim, o **frete médio** de cada loja foi analisado. As lojas com frete mais baixo podem ser mais atraentes para os clientes, já que o custo do frete muitas vezes influencia a decisão de compra.

- **Loja 1**: R$ 34,69

- **Loja 2**: R$ 33,62

- **Loja 3**: R$ 33,07

- **Loja 4**: R$ 31,28

A **Loja 4** tem o frete médio mais baixo, o que pode ser uma vantagem competitiva, especialmente para atrair mais clientes. O frete é um custo importante para os consumidores, e ter um custo mais baixo pode tornar a loja mais atraente, ajudando a aumentar as vendas.

---

## Conclusão

Após analisar todos os dados, a **Loja 3** parece ser a melhor opção para o Senhor João vender. Ela se destaca em vários aspectos importantes:

- Tem uma **boa média de avaliações** (4.05), o que indica que os clientes estão satisfeitos com os produtos e o atendimento.
- Apesar de não ser a loja com maior faturamento, ela tem **produtos mais vendidos** com boa aceitação no mercado.
- Seu **frete médio** é razoável e não é o mais alto, o que a torna competitiva no custo total para os clientes.

Portanto, considerando o equilíbrio entre faturamento, satisfação dos clientes e a competitividade no mercado e a maior probabilidade de uma venda rápida a **Loja 3** parece ser a escolha mais estratégica para venda.
